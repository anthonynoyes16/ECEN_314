```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as sig
import array as arr
import math
from sympy import *
```

```python
# ak
plt.figure(figsize=(8,4))
i = 0
ak = arr.array('f', [])
x = arr.array('i', [])
for i in range(101):
    if (i-50 == 0):
        ak.append(0.5)
        x.append(i-50)
        continue
    val = (math.sin((i-50)*math.pi*0.5))/((i-50)*math.pi)
    ak.append(val)
    x.append(i-50)

plt.plot(x, ak)
plt.title('ak')
plt.xlabel('k')
plt.ylabel('ak')
```

```python
# bk
plt.figure(figsize=(8,4))
i = 0
bk = arr.array('f', [])
for i in range(101):
    if (i-50 == 0):
        bk.append(0.5)
        continue
    val = (((-1*math.cos((i-50)*math.pi))+1)/(((i-50)**2)*(math.pi**2)))
    bk.append(val)
    
plt.plot(x, bk)
plt.title('bk')
plt.xlabel('k')
plt.ylabel('bk')
```

```python
n = 10
x_t = 0
x = arr.array('f', [])
t_arr = arr.array('f', [])
for t in range (481):
    x_t = 0
    for i in range (n+1):
        if (i == 0):
            x_t += 1/2
            continue
        x_t += 2 * ak[i+50] * math.cos(i*math.pi*(t-240)/80)
    x.append(x_t)
    t_arr.append((t-240)/80)

plt.plot(t_arr, x)
plt.title('x_t, n = 10')
plt.xlabel('t')
plt.ylabel('x_t')
```
