```python
from sympy import *
var('t')
y_t = Function('y')(t)
lccde = Eq(y_t.diff(t)+y_t, 0)
dsolve(lccde)
```

```python
y_t = Piecewise((0, t < 0), (exp(-t), t > 0))
plot(y_t, title = "Exercise 1 (iii)", xlabel = 't', ylabel = "y(t)")
```

```python
y_t = Piecewise((0, t < pi/2), (-exp(pi/2)/exp(t) + sin(t) - cos(t), t > pi/2))
plot(y_t, title = "Exercise 1 (iiii)", xlabel = 't', ylabel = "y(t)")
```
