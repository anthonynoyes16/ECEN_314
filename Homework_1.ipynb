```python
from sympy import *
var('t')
x_t = Piecewise((0, t < -1), (t + 1, (-1 <= t) & (t < 0)), (1, (0 <= t) & (t < 1)), (0, t >= 1))
plot(x_t, (t, -2, 2), title = "Exercise 1 (i)", xlabel = 't', ylabel = "x(t)")
```

```python
var('x')
y_1 = x_t.subs(t, -t)
# y1(t) is x(t) reflected across the y axis
y_2 = x_t.subs(t, t - 1)
# y2(t) is x(t) with the center shifted to the right 1 unit
y_3 = x_t.subs(t, t + 1)
# y3(t) is x(t) with the center shifted to the left 1 unit
y_4 = x_t.subs(t, 2 * t)
# y4(t) is x(t) shrunk down to a scale of 1/2 of the original
y_5 = x_t.subs(t, t / 2)
# y5(t) is x(t) expanded to twice the scale of the original
plot(y_1, (t, -2, 2), title = "Exercise 1 (ii) y1(t)", xlabel = 't', ylabel = "x(t)")
plot(y_2, (t, -2, 2), title = "Exercise 1 (ii) y2(t)", xlabel = 't', ylabel = "x(t)")
plot(y_3, (t, -2, 2), title = "Exercise 1 (ii) y3(t)", xlabel = 't', ylabel = "x(t)")
plot(y_4, (t, -2, 2), title = "Exercise 1 (ii) y4(t)", xlabel = 't', ylabel = "x(t)")
plot(y_5, (t, -2, 2), title = "Exercise 1 (ii) y5(t)", xlabel = 't', ylabel = "x(t)")
```
