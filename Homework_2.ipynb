```python
from sympy import *
var('t')
x_t1 = Piecewise((0, t <= -1), (1, (t > -1) & (t < 0)), (1/2, (t >= 0) & (t < 1)), (0, t >= 1))
plot(x_t1, (t, -1, 1), title = "Exercise 1 (ii)", xlabel = 't', ylabel = "x(t)")
plot(x_t1.subs(t, 1-t/2), (t, -4, 4), title = "Exercise 1 (ii)", xlabel = 't', ylabel = "x(1-t/2)")
```

```python
x_t2 = Piecewise((0, t <= -1), (t + 1, (t >= -1) & (t < 0)), (1, (t >= 0) & (t < 1)), (0, t >= 1))
plot(x_t2, (t, -1, 1), title = "Exercise 2 (i)", xlabel = 't', ylabel = "x(t)")
plot((x_t2 + x_t2.subs(t, -t)) / 2, (t, -1, 1), title = "Exercise 2 (ii)", xlabel = 't', ylabel = "xe(t)")
plot((x_t2 - x_t2.subs(t, -t)) / 2, (t, -1, 1), title = "Exercise 2 (ii)", xlabel = 't', ylabel = "xo(t)")
```
